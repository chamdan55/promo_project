```python
def generate_random_date(start_date, end_date):
    # Generate a random date between start_date and end_date
    delta = end_date - start_date
    random_days = random.randint(0, delta.days)
    random_date = start_date + timedelta(days=random_days)
    return random_date.strftime('%Y-%m-%d %H:%M:%S %z')  # Format to 'DD/MM/YY'

def generate_random_date2(start_date, end_date):
    # Generate a random date between start_date and end_date
    delta = end_date - start_date
    random_minute = random.randint(0, int(delta.total_seconds() // 60))
    random_date = start_date + timedelta(minutes=random_minute)
    return random_date.strftime('%Y-%m-%d %H:%M:%S %z')
```
