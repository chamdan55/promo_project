```python
import pandas as pd
import random
import numpy as np
import nanoid
import itertools
from datetime import datetime, timedelta

random.seed(42)
```

```python
def generate_random_date(start_date, end_date):
    # Generate a random date between start_date and end_date
    delta = end_date - start_date
    random_days = random.randint(0, delta.days)
    random_date = start_date + timedelta(days=random_days)
    return random_date.strftime('%d/%m/%y')  # Format to 'DD/MM/YY'
```

```python
data = {
    'MCC_GROUP': ['ALFAMART', 'INDOMARET', 'SUPERINDO', 'ERAFONE', 'IBOX', 'BARAYA', 'DAYTRANS', 'GRAPARI', 'GERAI IM3', 'IKEA', 'DECATHLON'],
    'MERCHANT_CATEGORY': ['Food & Beverage', 'Lifestyle', 'Travel & Transportation', 'Gadget & Electronics', 'E-Commerce', 'Health & Beauty', 'Others/Miscellaneous', 'Telco'],
    'ASSETS': list(range(1, 6)),
    'AGE': list(range(1, 6)),
    'SEGMENTATION': ['All Segment', 'Working Professionals', 'Emerging Wealth', 'Wealthy Achievers', 'Young Gen & Early Jobber', 'Mature High Networth', 'Silver Ager', 'Minimal'],
    'RELIGION': ['Islam', 'Christian', 'Catholic', 'Buddha', 'Hindu', 'Konghucu', 'Others'],
    'GENDER': ['M', 'F'],
    'LOCATION': ['Nasional', 'Balikpapan', 'Bandung', 'Denpasar', 'Jakarta', 'Makassar', 'Malang', 'Manado', 'Medan', 'Padang', 'Palembang', 'Papua', 'Semarang', 'Surabaya', 'Yogyakarta'],
    'TIME': ['WEEKDAY - WORKING TIME', 'WEEKDAY - RESTING TIME', 'WEEKDAY - TRAVELING TIME', 'WEEKEND - WEEKEND'],
    'RANKING': list(range(1,11))
}
```

```python
# Generate all combinations of list1 and list2
combinations = list(itertools.product(data['MCC_GROUP'], data['MERCHANT_CATEGORY']))

master_merchant = pd.DataFrame(combinations, columns=['MCC_GROUP', 'MERCHANT_CATEGORY'])
n = len(master_merchant)
times = datetime.now().strftime('%Y-%m-%d %H:%M:%S %z')
times
other = {
    'ID': [str(nanoid.generate(size=21)) for _ in range(n)],
    'UPDATE_BY': ['Camdun' for _ in range(n)],
    'UPDATED_TIME': [times for _ in range(n)],
    'CREATED_BY': ['Camdun' for _ in range(n)],
    'CREATED_TIME': [times for _ in range(n)]
}

for key, val in other.items():
    master_merchant[key] = other[key]

master_merchant = master_merchant[['ID', 'MCC_GROUP', 'MERCHANT_CATEGORY', 'UPDATE_BY', 'UPDATED_TIME', 'CREATED_BY', 'CREATED_TIME']]
master_merchant
```
